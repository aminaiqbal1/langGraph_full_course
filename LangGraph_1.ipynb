# **1. What is LangGraph?**

LangGraph ek library hai jo LangChain ecosystem ka hissa hai. Yeh library multiple LLM agents (ya chains) ko ek system mein define, manage aur execute karne ka asaan tareeqa deti hai.

Example:
Aap ek AI system bana rahe hain jo multiple tasks karta hai, jaise user input lena, process karna aur output generate karna. LangGraph is process ko simplify kar deta hai aur workflows ko organized banata hai.

2. Why is it important to create a graph in LangGraph?
LangGraph mein graph ek complex workflow ka structure represent karta hai. Har node ek specific kaam represent karti hai, aur edges batati hain ke yeh kaam kaise ek dusre se connected hain.
Example:
Agar ek chatbot design karna hai jo user se input leta hai, process karta hai, aur jawab generate karta hai, to graph yeh batata hai ke yeh steps kaise linked hain aur smoothly kaise kaam karenge.

3. What is a State?
State ka matlab hai system ke liye "yaad rakhne ki salahiyat." Agar state na ho, to system har request ko independent treat karega aur context bhool jayega.
Example:
Aap poochte hain "Mujhe kal ka weather batao," aur phir "Aaj ka weather kya hai?" Agar state manage nahi ki gayi, to system aapka pehla sawaal bhool jayega. State ki wajah se system context ko yaad rakhta hai aur relevant jawab deta hai.

4. What is a Node?
Node ek individual step ya unit hai jo workflow ka hissa hota hai. Har node ek specific task perform karta hai aur apne output ko agle node tak bhejta hai.
Example:
Ek AI system jo restaurant recommendations deta hai:
Node 1: User ka input lena (e.g., "Chinese food chahiye").
Node 2: Database se options retrieve karna.
Node 3: Best option recommend karna.
Har node ek step perform karta hai aur connected hota hai.

5. What is an Edge?
Edge do nodes ke beech ka connection hota hai. Yeh batata hai ke ek node ka output agle node tak kaise flow karega.
Example:
Agar ek chatbot mein:
Node 1: Input lena.
Node 2: Analyze karna.
Node 3: Response generate karna.
Edges yeh flow define karti hain, jaise input se analysis tak aur analysis se response tak.

6. What is a Conditional Edge?
Conditional edge ek special edge hai jo conditions pe kaam karti hai. Yeh decide karti hai ke agla step kya hoga, based on a specific condition ya result.
Example:
Agar ek recommendation system poochta hai:
"Veg chahiye ya non-veg?"
Agar user "veg" kahe, to Node A activate hoga.
Agar user "non-veg" kahe, to Node B activate hoga.
Yeh edge conditions ke basis par path decide karti hai.

7. What is Tool Binding with LLMs?
Tool binding ka matlab hai LLMs ko external tools (APIs, databases, ya functions) ke sath connect karna taake yeh real-world tasks perform kar saken.
Example:
Agar ek tool "fetch_weather" bind kiya gaya hai, to LLM query "Aaj ka weather kya hai?" ka jawab real-time data fetch kar ke dega.
8. What are Reducers in LangGraph?
Reducers wo functions hain jo graph ki state ko update karte hain aur multiple nodes se aane wale data ko combine karte hain.
Example:
Agar ek graph ke do nodes weather aur traffic ka data de rahe hain, to reducer in dono outputs ko merge karke ek consistent state banata hai jo agle steps ke liye useful hoti hai.

# **Overview¶**
# **LangGraph**

 is a library for building stateful, multi-actor applications with LLMs, used to create agent and multi-agent workflows. Compared to other LLM frameworks, it offers these core benefits: cycles, controllability, and persistence. LangGraph allows you to define flows that involve cycles, essential for most agentic architectures, differentiating it from DAG-based solutions. As a very low-level framework, it provides fine-grained control over both the flow and state of your application, crucial for creating reliable agents. Additionally, LangGraph includes built-in persistence, enabling advanced human-in-the-loop and memory features.

LangGraph is inspired by Pregel and Apache Beam. The public interface draws inspiration from NetworkX. LangGraph is built by LangChain Inc, the creators of LangChain, but can be used without LangChain.

LangGraph Platform is infrastructure for deploying LangGraph agents. It is a commercial solution for deploying agentic applications to production, built on the open-source LangGraph framework. The LangGraph Platform consists of several components that work together to support the development, deployment, debugging, and monitoring of LangGraph applications: LangGraph Server (APIs), LangGraph SDKs (clients for the APIs), LangGraph CLI (command line tool for building the server), LangGraph Studio (UI/debugger),



To learn more about LangGraph, check out our first LangChain Academy course, Introduction to LangGraph, available for free here.

# **Key Features¶**
Cycles and Branching: Implement loops and conditionals in your apps.


Persistence: Automatically save state after each step in the graph. Pause and resume the graph execution at any point to support error recovery, human-in-the-loop workflows, time travel and more.


Human-in-the-Loop: Interrupt graph execution to approve or edit next action planned by the agent.


Streaming Support: Stream outputs as they are produced by each node (including token streaming).

Integration with LangChain: LangGraph integrates seamlessly with LangChain and LangSmith (but does not require them).

# **LangGraph Platform¶**
LangGraph Platform is a commercial solution for deploying agentic applications to production, built on the open-source LangGraph framework. Here are some common issues that arise in complex deployments, which LangGraph Platform addresses:

Streaming support: LangGraph Server provides multiple streaming modes optimized for various application needs
Background runs: Runs agents asynchronously in the background
Support for long running agents: Infrastructure that can handle long running processes
Double texting: Handle the case where you get two messages from the user before the agent can respond
Handle burstiness: Task queue for ensuring requests are handled consistently without loss, even under heavy loads

## Context

At LangChain, we aim to make it easy to build LLM applications. One type of LLM application you can build is an agent. There’s a lot of excitement around building agents because they can automate a wide range of tasks that were previously impossible.

In practice though, it is incredibly difficult to build systems that reliably execute on these tasks. As we’ve worked with our users to put agents into production, we’ve learned that more control is often necessary. You might need an agent to always call a specific tool first or use different prompts based on its state.

To tackle this problem, we’ve built [LangGraph](https://langchain-ai.github.io/langgraph/) — a framework for building agent and multi-agent applications. Separate from the LangChain package, LangGraph’s core design philosophy is to help developers add better precision and control into agent workflows, suitable for the complexity of real-world systems.
